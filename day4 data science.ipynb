```python
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second level')
plt.plot(x,y, marker='o',color='blue',mfc='r');
```

```python
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second level')
plt.plot(x,y, marker='o',color='red',mfc='r');
plt.grid()
#marker => <,>,^,0,*,H
```

```python
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second level')
plt.plot(x,y, marker='o',color='blue',mfc='r',ls='--',ms=9);
plt.grid()
#marker => <,>,^,0,*,H
```

```python
plt.subplot(2,2,1)
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second level')
plt.plot(x,y, marker='*',color='red',mfc='black',ls ='--',ms=9);
plt.plot(x,z, marker='o',color='red',mfc='blue',ls ='--',ms=9);
plt.grid()
```
